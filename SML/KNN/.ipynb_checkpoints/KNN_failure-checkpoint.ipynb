# KNN Failure

In scenarios of KNN = 3, there will always get 2/3 votes for the wrong class. This is because the closest point to the unknown point is itself and next are two other classes.

```python
import import_ipynb
import numpy as np
import matplotlib.pyplot as plt
from KNN_Intuition import KNN
```

The data to be generated is a 2-D 64-datapoint grid with 8 rows and 8 columns. 

Change target after going to the next point such that if t = 1 then it becomes 0 and vice versa. Same thing is done to the start target (start_t)

```python
def get_data():
    width = 8
    height = 8
    N = width * height
    X = np.zeros((N,2))
    Y = np.zeros(N)
    n = 0
    start_t = 0 # start target is 0 and will alternate between 0 and 1. 
    
    # loop through rows and columns to fill in values
    for i in range(width):
        t = start_t
        for j in range(height): 
            X[n] = [i, j]
            Y[n] = t
            n+=1
            t = (t + 1 )%2 
        start_t = (start_t + 1)%2
    return X,Y
```

Main function is then defined to get the data and plot the scatter plot to visualise it. 

The model is then created for K = 3 and fit in order to retrieve the training accuracy

```python
if __name__ == '__main__':
    X,Y = get_data()
    
    plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha=0.5 )
    plt.show()
    
    model = KNN(3)
    model.fit(X,Y)
    
    print("Train accuracy", model.score(X,Y))
```
